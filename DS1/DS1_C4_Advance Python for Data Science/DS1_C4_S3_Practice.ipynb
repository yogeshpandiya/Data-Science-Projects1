```python
Rural=Laon_Data[['Loan_ID','ApplicantIncome']][(Laon_Data['Property_Area']=='Rural')]
A=(Rural.rename(columns={'ApplicantIncome':'Rural Applicant Income','Loan_ID':'R_Loan_ID'})).set_index(np.arange(0,179))

Semiurban=Laon_Data[['Loan_ID','ApplicantIncome']][(Laon_Data['Property_Area']=='Semiurban')]
B=(Semiurban.rename(columns={'ApplicantIncome':'Semiurban Applicant Income','Loan_ID':'S_Loan_ID'})).set_index(np.arange(0,233))

Urban=Laon_Data[['Loan_ID','ApplicantIncome']][(Laon_Data['Property_Area']=='Urban')]
C=(Urban.rename(columns={'ApplicantIncome':'Urban Applicant Income','Loan_ID':'U_Loan_ID'})).set_index(np.arange(0,202))


Final_Data=pd.concat([A,B,C],axis=1)
Final_Data.sample(50)
```
