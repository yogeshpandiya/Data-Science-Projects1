# TASK1

### Establish the null and alternate hypothesis

H0: the average score in math is lesser or equal to 71 i.e. null hypothesis

Ha: the average score in math is more than 71.

The test needs to be conducted for the value which is greater than the population mean. 

So, we have to perform the right-tailed test.

```python
sample_data = [61,93,75,67,84,95]

#sample_data = [100,93,100,100,100,100] # to validate the hypothesis
```

### Calulating Mean and Standard Deviation

```python
import statistics as st

s_mean=st.mean(sample_data)
s_sd=st.stdev(sample_data)
n=len(sample_data)
print('Sample Mean = '+str(s_mean))
print('Sample Standard Deviation = '+str(s_sd))
print('Sample Size = '+str(n))
```

### Determine the appropriate statistical test

```python
p_mean = 71 #Population Mean
s_mean =  79.16666666666667 #  Sample Mean
s_sd = 13.862419221285535 #  Sample Standard Deviation
n = 6 # Sample Size
dof = n-1 # Degrees of Freedom
#We will test hypothesis about a population mean using the t-statistic (standard deviation of population is unknown).
```

### Set the value of alpha

```python
alpha = 0.05 # Significance level
```

### Establish the decision rule

i. If p_value < alpha : Rejection of Null Hypothesis (H0)

ii. If -t-critical > t-statistic > +t-critical : Rejection of Null Hypothesis(H0)

### Analyze the data

```python
import math
from scipy.stats import t

t_statistic = (s_mean - p_mean) / (s_sd/math.sqrt(n)) 
print("The t-statistic is ", t_statistic)

p_value=t.sf(abs(t_statistic),dof)
print("The p_value is ", p_value)

t_critical = t.ppf(1-alpha, dof)
print("The t-critical value is "+str(t_critical))
```

### Reach a statistical conclusion

#As per, decision rule:

i. p_value < alpha : Rejection of Null Hypothesis (H0) # p_value < alpha : Rejection of Null Hypothesis (H0) while considering both the tails

ii. If -t-critical > t-statistic > +t-critical : Rejection of Null Hypothesis(H0)

In our study,

p_value=0.10429845495147672 > alpha= 0.05, which leads to acceptance of Null Hypothesis

t_statistic=1.443050156931714 < t-critical=2.015048372669157, which leads to leads to acceptance of Null Hypothesis.

### Make a business decision

In statistical conclusion our null hypothesis is accepted which lead us to make this descision that the average score in math is lesser or equal to 71.

```python
import matplotlib.pyplot as plt
from scipy.stats import t
import numpy as np


upper=s_mean+4*s_sd
lower=s_mean-4*s_sd

x=np.arange(lower,upper,.005)

tvals17=t.pdf(x,5)


plt.plot(x,tvals17,color='brown')
plt.legend(['df=17'])

plt.show()
```

# TASK2

### Establish the null and alternate hypothesis

H0: the difference between two sample means is zero.

Ha: the difference between two sample means are not equal to zero.

### Data

```python
Machine1 = [4.05,4.01,4.04,4.02]
Machine2 = [4.02, 3.99, 4.01, 3.99, 4.00, 4.00]
#Machine2 = [1, 1, 4, 1, 1, 1] # to validate the hypothesis
```

### Determine the appropriate statistical test

 We will test hypothesis about a population mean using the t-statistics (standard deviation of population is known).

t-statistic = [(𝑥1- 𝑥2) – (µ1 - µ2)] / [sqrt((s1^2*(n1 -1) + s2^2*(n2-1))/df)*sqrt(1/n1 + 1/n2)]
Where, df is the degrees of freedom.


### Set the value of alpha (α)

It is given that a 99% level of confidence to be used to test hypothesis.

so significance leve would be 100-99% e.i = 1% is our significance level

hence

alpha (α) = 0.01

This test is a two-tailed test, each of the two rejection regions has an area of 0.005


### Establish the decision rule

i. If p_value < alpha : Rejection of Null Hypothesis (H0)

ii. If -t-critical > t-statistic > +t-critical : Rejection of Null Hypothesis(H0)

### Set degrees of freedom


Sample size machine one (n1)=4

Sample size of machine two (n2)= 6

Degrees of Freedom (df) = n1 + n2 - 2 = 4 + 6 - 2 = 8

### Analyze the data

```python
import statistics as st
x1=st.mean(Machine1)
x2=st.mean(Machine2)
v1=(st.stdev(Machine1))**2
v2=(st.stdev(Machine2))**2
n1=len(Machine1)
n2=len(Machine2)
df= n1+n2-2
alpha = 0.01/2
print("Sample mean of Machine1 =",x1)
print("Sample mean of Machine2 =",x2)
print("Sample variance of Machine1 =",v1)
print("Sample variance of Machine2 =",v2)
print("Sample size of Machine1 =",n1)
print("Sample size of Machine2 =",n2)
print('Degrees of freedom =',df)
print('Value of alpha is =',alpha)
```

### Calculate t-statistic and t-critical

```python
import math
from scipy.stats import t
signal = abs(x1-x2)
noise = math.sqrt((v1*(n1-1) + v2*(n2-1))/df)*math.sqrt(1/n1 + 1/n2)
t_statistic=signal/noise
print('t-statistic =',t_statistic)

p_value=t.sf(abs(t_statistic),df)*2
print("The p_value is ", p_value)

t_critical = t.ppf(1-alpha, df)
print('t-critical =',t_critical)
```

### Statistical Conclusion

As per, decision rule:

i. p_value < alpha : Rejection of Null Hypothesis (H0)

ii. If -t-critical > t-statistic > +t-critical : Rejection of Null Hypothesis(H0)

In our study,

p_value=0.016409957941527966 > alpha=0.01, which leads to acceptance of Null Hypothesis

t_statistic= 3.0259602845318176< t-critical=3.3553873313333957, which leads to acceptance of Null Hypothesis.

### Business Descision

With the help of statical conclusion we have accepted the null hypothesis which mean, there is no difference in average fill of these two machines

```python
x1=st.mean(Machine1)
x2=st.mean(Machine2)
v1=st.stdev(Machine1)
v2=st.stdev(Machine2)


upper=x1+4*v1
lower=x1-4*v1
upper2=x2+4*v2
lower2=x2-4*v2

x=np.arange(lower,upper,.05)
y=np.arange(lower2,upper2,.05)

tvals1=t.pdf(x,4)
tvals2=t.pdf(x,6)


plt.plot(x,tvals1,color='brown')
plt.plot(x,tvals2,color='green')
plt.legend(['df=4','df-6'])

plt.show()
```

# TASK3

### Establishing the null and alternate hypothesis


H0: the difference between sample means  of machine is zero.

Ha: the difference between sample means are not equal to zero.

### Establish the decision rule

i. p_value < alpha : Rejection of Null Hypothesis (H0)

ii. If -t-critical > t-statistic > +t-critical : Rejection of Null Hypothesis(H0)

### Access the sample data

```python
Machine1 = [4.05,4.01,4.04,4.02]
Machine2 = [4.02, 3.99, 4.01, 3.99, 4.00, 4.00]
Machine3 = [4.00,3.97,3.95,3.98,3.97]
Machine4 = [4.01,4.00,3.99,4.02]
```

### Calculate the total for each sample(Tj)

```python
T1=sum(Machine1)
T2=sum(Machine2)
T3=sum(Machine3)
T4 = sum(Machine4)
T=T1+T2+T3+T4
print('Total from Machine1 =',T1 )
print('Total from Machine2 =',T2 )
print('Total from Machine3 =',T3 )
print('Total from Machine4 =',T4 )
print('Total from all 4 Machine =',T )
```

### Samples size (nj)

```python
n1=len(Machine1)
n2=len(Machine2)
n3=len(Machine3)
n4=len(Machine4)
n=n1+n2+n3+n4
print('Size of Machine1 =',n1)
print('Size of Machine2 =',n2)
print('Size of Machine3 =',n3)
print('Size of Machine4 =',n4)
print('Size of all samples =',n)
```

### Calculate sample means (xj)``

```python
import statistics as st

x1=st.mean(Machine1)
x2=st.mean(Machine2)
x3=st.mean(Machine3)
x4=st.mean(Machine4)
x=(x1+x2+x3+x4)/4
print('Mean of the Machine 1 =',x1)
print('Mean of the Machine 2 =',x2)
print('Mean of the Machine 3 =',x3)
print('Mean of the Machine 4 =',x4)
print('Mean of the all samples =',x)
```

### Analyze the data

##### Calculate sum of squares columns (SSC)

```python
SSC=n1*(x1-x)**2 + n2*(x2-x)**2 + n3*(x3-x)**2 + n4*(x4-x)**2
print("SSC =",SSC)
```

#### Calculate sum of squares errors (SSE)

```python
SSE=0
for i in Machine1:
    SSE=SSE+(i-x1)**2
    
for i in Machine2:
    SSE=SSE+(i-x2)**2
    
for i in Machine3:
    SSE=SSE+(i-x3)**2
    
for i in Machine4:
    SSE=SSE+(i-x4)**2
    
print("SSE =",SSE)
```

#### Calculate total sum of squares (SST)


```python
SST=0
for i in Machine1:
    SST=SST+(i-x)**2
    
for i in Machine2:
    SST=SST+(i-x)**2
    
for i in Machine3:
    SST=SST+(i-x)**2

for i in Machine4:
    SST=SST+(i-x)**2
    
print("SST =",SST)
```

#### Calculate column, error and total degree of freedoms (dfc,dfe,dft)

```python
dfc= 4-1 #column -1
dfe= 19-4 
dft= 19-1
print('dfc =',dfc)
print('dfe =',dfe)
print('dft =',dft)
```

#### Calculate MSC

```python
MSC=SSC/dfc
print('MSC =',MSC)
```

#### Calculate MSE

```python
MSE=SSE/dfe
print('MSE =',MSE)
```

### Calculate F-statistic

```python
f_statistic=MSC/MSE
print('F-statistic =',f_statistic)
```

### Calculate F-critical


```python
from scipy.stats import f

f_critical=f.ppf(1-0.01, dfc, dfe)
print('F-critical =',f_critical)
print('F-statistic =',f_statistic)

p_value=f.sf(abs(f_statistic),dfc,dfe)
print("The p_value is ", p_value)
print('Value of alpha is 0.01')
```

### Statistical Conclusion

As per, decision rule:

i. If p_value < alpha : Rejection of Null Hypothesis (H0)

ii. If F-statistic > F-critical : Rejection of Null Hypothesis(H0)

In our study,

p_value=0.0006623957200780382 < alpha=0.01, which leads to rejection of Null Hypothesis

F-statistic=10.169045353631299> F-critical=5.41696485781842, which leads to rejection of Null Hypothesis.

### Business Decision

##### *In our Statical Conclusion Null Hypothesis got rejected hence there is a difference in machine average*
