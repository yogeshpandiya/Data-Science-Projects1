## Establish the null and alternate hypothesis

H0: the average rent for a special purpose industiral space is 33.88 acres in 2020.

Ha: at present the average rent for a special purpose industiral space is not 33.88.

The test needs to be conducted for the value which is greater than the population mean. 


The test needs to be conducted for the value which is greater than or less than the difference of population means. So, we have to perform the two-tailed test.

## Determine the appropriate statistical test

```python
mu = 33.88 #population mean
sd_sigma =1.28  #population standard deviation 
alpha = 0.5 #value of alpha
n = 49 #sample_size
x_bar = 32.18 #sample mean 
```

## Establish the decision rule

i. If p-value < alpha : Rejection of Null Hypothesis(H0)
    
ii. If -z-critical > z-statistics > +z-critical : Rejection of Null Hypothesis(H0)

## Analyze the data

```python
import math
from scipy.stats import norm
```

```python
z_stat = (x_bar - mu)/(sd_sigma/math.sqrt(n))
p_values = norm.sf(abs(z_stat))*2
z_critical = norm.ppf(1-alpha/2)
```

```python
print("The Z statistics is ", z_stat)
print("The p_value is "+str(p_values))
print("The z-critical value is "+str(z_critical))
```

## Reach a statistical conclusion

As per, decision rule:

i. If p-value < alpha : Rejection of Null Hypothesis(H0)  

ii. If -z-critical > z-statistics > +z-critical : Rejection of Null Hypothesis(H0)

In our study,

p_value=1.4463410868771357e-20 < alpha=0.025, which leads to reject the Null Hypothesis.

z_statistics=-9.296875000000016 < z-critical=-0.6744897501960817, which leads to reject of Null Hypothesis.

## Make a business decision

There is a change in avarage rent of space at present the average rent for a special purpose industiral space is not 33.88.
