# TASK1

```python
DATA= pd.read_csv('DS1_C5_S1_Descriptive_Analysis_Practice_Car_Data.csv')
```

```python
Data = DATA[['Make','Model','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Minimum_Turning_Radius']].dropna()
```

```python
Data1 = Data[['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Minimum_Turning_Radius']]
Mean=[]
Median=[]
SD=[]
CV=[]

for col in Data1:
    col_mean=Data1.loc[:,col].mean() 
    Mean.append(col_mean)
    Median.append(Data1.loc[:,col].median()) 
    col_sd=Data1.loc[:,col].std()
    SD.append(col_sd)
    CV.append(col_sd/col_mean*100)
    
row_head=['Mean','Median','SD','CV']
col_names=['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Minimum_Turning_Radius']

d_data=pd.DataFrame([Mean, Median, SD,CV],columns=col_names)
d_data.insert(0, "Measures", row_head, True)

d_data
```

```python
A= pd.pivot_table(Data, index=['Make'])
A.head()
```

```python
for i in A.index:
    Data1 = Data[['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Minimum_Turning_Radius']][Data['Make']==i]
    Mean=[]
    Median=[]
    SD=[]
    CV=[]

    print('\n','--------------------------___{0}___------------------------------'.format(i),'\n')
    for col in Data1:
        col_mean=Data1.loc[:,col].mean() 
        Mean.append(col_mean)
        Median.append(Data1.loc[:,col].median()) 
        col_sd=Data1.loc[:,col].std()
        SD.append(col_sd)
        CV.append(col_sd/col_mean*100)

    row_head=['Mean','Median','SD','CV']
    col_names=['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Minimum_Turning_Radius']

    d_data=pd.DataFrame([Mean, Median, SD,CV],columns=col_names)
    d_data.insert(0, "Measures", row_head, True)
    print(d_data)

```

```python
Data_2 = DATA[['City_Mileage_km_litre','Displacement','Cylinders','Valves_Per_Cylinder','Fuel_Tank_Capacity_litre','Height_mm','Length_mm','Width_mm','Wheelbase','Minimum_Turning_Radius']]
Data2 = Data_2.dropna()
Data2
```

```python
fig, ((ax1, ax2),(ax3, ax4))=plt.subplots(nrows=2,ncols=2,figsize=(20,10))
fig.suptitle('Coefficient Correlation graphs', fontsize=20 )
ax1.scatter(Data2['City_Mileage_km_litre'],Data2['Displacement'])
ax1.title.set_text('Displacement')
ax2.scatter(Data2['City_Mileage_km_litre'],Data2['Cylinders'])
ax2.title.set_text('Cylinders')
ax3.scatter(Data2['City_Mileage_km_litre'],Data2['Fuel_Tank_Capacity_litre'])
ax3.title.set_text('Fuel_Tank_Capacity_litre')
ax4.scatter(Data2['City_Mileage_km_litre'],Data2['Width_mm'])
ax4.title.set_text('Width_mm')
plt.show()
```

```python
mc_corr_D= Pearsons_Correlation(Data2['City_Mileage_km_litre'],Data2['Displacement'])
mc_corr_C=Pearsons_Correlation(Data2['City_Mileage_km_litre'],Data2['Cylinders'])
mc_corr_F=Pearsons_Correlation(Data2['City_Mileage_km_litre'],Data2['Fuel_Tank_Capacity_litre'])
mc_corr_W=Pearsons_Correlation(Data2['City_Mileage_km_litre'],Data2['Width_mm'])
print('\n','Coefficient Correlation between City_Mileage_km_litre and Displacement is:',mc_corr_D)
print('\n','Coefficient Correlation between City_Mileage_km_litre and Cylinders is:',mc_corr_C)
print('\n','Coefficient Correlation between City_Mileage_km_litre and Fuel_Tank_Capacity_litre is:',mc_corr_F)
print('\n','Coefficient Correlation between City_Mileage_km_litre and Width_mm is:',mc_corr_W)
```
