# Stratified Sampling

```python
grouped_d_df=Data.groupby(['Year'])
dep=Data.Year.unique()
Strat_Sample_df=pd.DataFrame()
for d in dep:
    d_data=grouped_d_df.get_group(d)
    sample_data=d_data.sample(frac=.3,replace=True, random_state=1)
    Strat_Sample_df=Strat_Sample_df.append(sample_data, ignore_index=True)
    
Strat_Sample_df
```

```python
grouped_d_df=Data.groupby(['Genre'])
dep=Data.Genre.unique()
Strat_Sample_df=pd.DataFrame()
for d in dep:
    d_data=grouped_d_df.get_group(d)
    sample_data=d_data.sample(frac=.3,replace=True, random_state=1)
    Strat_Sample_df=Strat_Sample_df.append(sample_data, ignore_index=True)
    
Strat_Sample_df
```
