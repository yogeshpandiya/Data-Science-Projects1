```python
from decimal import Decimal
def The_Conversion_App(value):    
    
    if type(value) != str: 
        Radian_to_Degree = "{0}{1} Degree".format(round(57.296*value,2),chr(176)) #used .fromat and round function to format the value
        
    else: 
        Radian_to_Degree = 'Radian Can not be string'
#---------------------##################-------------------------------

    if type(value) != str:
        Degree_to_Radian = "{0} Radian".format(round(0.01745*value,2)) 
                                                     
    else: 
        Degree_to_Radian = 'Degree Can not be string'
#---------------------##################-------------------------------

    if value==0:
        Number_to_factorial=1
    elif type(value) != float and type(value) != str:
        
        fact = 1
        for i in range(value):
            fact*=i+1
        Number_to_factorial = "Factorial of {0} is {1}".format(value,fact)
    else: 
        Number_to_factorial = 'Please enter Numeric Value'
#---------------------##################-------------------------------

    if type(value) != str:
        Number_to_Sq = value**0.5
    else: 
        Number_to_Sq = 'Please enter Numeric Value'
#---------------------##################-------------------------------

    if type(value) != str:
        a = 100000000*1
        Megabyte_to_Nible = "{:.2E}".format(Decimal(a))
    else: 
        Megabyte_to_Nible = 'Please enter Numeric Value'
#---------------------##################-------------------------------

    if type(value) == float:
        Fraction_to_Decimal = 'Decimal value is {0}'.format(value)
    else:
        Fraction_to_Decimal = 'Enter Correct Friction Number'
#---------------------##################-------------------------------

    if type(value) != str:
        klomtrPerh_to_MtrperSec = '{0} Kilometer Per Hour to Meter Per Second is {1}m/s'.format(value,round(0.277778*value,2))
    else:
        klomtrPerh_to_MtrperSec = 'Enter Correct Kilometer Per Hours'
        
        
#---------------------##################-------------------------------

    if type(value) != str:
        MtrperSec_to_klomtrPerh = '{0} Meter Per Second to Kilometer Per Hour is {1}Km/h'.format(value,round(3.6*value,2))
    else:
        MtrperSec_to_klomtrPerh = 'Enter Correct Kilometer Per Hours'

#---------------------##################-------------------------------

    if type(value) == int:
        int_to_float = 'Flaot Value of {0} is {1}'.format(value,float(value))
    else:
        int_to_float = 'Please provide Int value'
        
#---------------------##################-------------------------------

    if type(value) == float:
        float_to_int = 'Int Value of {0} is {1}'.format(value,int(value))
    else:
        float_to_int = 'Please provide float value'
        
#---------------------##################-------------------------------

    if type(value) == int:
        decimal_to_Octal = 'Oct Value of {0} is {1}'.format(value,oct(value))
    else:
        decimal_to_Octal = 'Please provide correct value'
#---------------------##################-------------------------------

    if type(value) == int:
        a = oct(value)
        Octal_to_decimal='Decimal Value of {0} is {1}'.format(value,int(str(a),8)) 
    else:
        Octal_to_decimal= 'Please provide correct adasd value'
        
#---------------------##################-------------------------------

    if type(value) == int:
        decimal_to_hex ='Hexadecimal Value of decimal{0} is {1}'.format(value, hex(value)) 
    else:
        decimal_to_hex= 'Please provide correct value'
#---------------------##################-------------------------------

    if type(value) == str:
        for ch in value:
            if ((ch < '0' or ch > '9') and
                (ch < 'A' or ch > 'F')): 
                pass
        hex_to_decimal = 'Decimal value of Hexadecimal {0} is {1}'.format(value, int(str(value),16))  
    else:
        hex_to_decimal= 'Please provide correct value'
        
#---------------------##################-------------------------------

    if type(value) == int:
        decimal_to_bin = 'Binaray value of decimal {0} is {1}'.format(value, bin(value)[2:])
    else:
        decimal_to_bin= 'Please provide correct value'
#---------------------##################-------------------------------

    if type(value) == str and value.count('0')+value.count('1')==len(value):
        bin_to_decimal = 'Decimal value of {0} is {1}'.format(value, int(value,2))
    else:
        bin_to_decimal= 'Please provide correct value'
        
#---------------------##################-------------------------------


    if  type(value)==int:
        ascii_to_chr = 'Character to Ascii of {0} is {1} '.format(value, chr(value)) 
    else:
        ascii_to_chr= 'Please provide correct value'

#---------------------##################-------------------------------

    if  type(value)==str:
        for i in value:
            if i.isascii():
                chr_to_ascii = ord(i)
    else:
        chr_to_ascii= 'Please provide correct value'

#---------------------##################-------------------------------

    if  type(value)==float or  type(value)==int:
        cel_to_far= "{0}{1} Farenheit".format(round((value*9/5)+32,2),chr(176))
    else:
        cel_to_far= 'Please provide correct value'

#---------------------##################-------------------------------

    
    if  type(value)==float or  type(value)==int:
        far_to_cal= "{0}{1} Celsius".format(round((value-32)*1.8,2),chr(176))
        
    else:
        far_to_cal= 'Please provide correct value'
        
        
    return Radian_to_Degree, Degree_to_Radian, Number_to_factorial,Number_to_Sq,Megabyte_to_Nible,Fraction_to_Decimal,klomtrPerh_to_MtrperSec,MtrperSec_to_klomtrPerh, int_to_float, float_to_int,decimal_to_Octal,Octal_to_decimal,decimal_to_hex,hex_to_decimal,decimal_to_bin,bin_to_decimal,ascii_to_chr,chr_to_ascii,cel_to_far,far_to_cal

Radian_to_Degree, Degree_to_Radian, Number_to_factorial,Number_to_Sq,Megabyte_to_Nible,Fraction_to_Decimal,klomtrPerh_to_MtrperSec,MtrperSec_to_klomtrPerh, int_to_float, float_to_int,decimal_to_Octal,Octal_to_decimal,decimal_to_hex,hex_to_decimal,decimal_to_bin,bin_to_decimal,ascii_to_chr,chr_to_ascii,cel_to_far,far_to_cal= The_Conversion_App(int(input('Enter a Value  ')))
```
