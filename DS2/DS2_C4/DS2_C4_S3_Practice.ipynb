```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

# TASK1

```python
df1 = pd.read_csv('DS2_C6_S3_CarSales_Data_Practice.csv')
```

```python
df1.drop(['__year_resale_value','Latest_Launch'],axis=1, inplace = True)
df1.dropna(inplace = True)
df=df1.copy()
df.head()
```

# TASK2

```python
X=df[['Price_in_thousands','Power_perf_factor']]
plt.figure(figsize=(30, 10))
dendrogrm = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.axhline(y=150, color='r', linestyle='--')
ax = plt.gca()
ax.tick_params(axis='x', which='major', labelsize=10)
plt.title('Price Dendrogram',fontsize=30)
plt.xlabel('Model',fontsize=20)
plt.ylabel('Euclidean distance', fontsize=20)
plt.show()
```

# TASK3

```python
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'complete')
y_hc = hc.fit_predict(X)
```

```python
df.head()
```

```python
X = np.array(X)
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 50, c = 'red')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 50, c = 'blue')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 50, c = 'green')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 50, c = 'cyan')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 50, c = 'magenta')
plt.title('Clusters')
plt.xlabel('Price_in_thousands')
plt.ylabel('Power_perf_factor')
plt.show()
```
