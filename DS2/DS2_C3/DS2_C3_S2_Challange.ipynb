```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

from sklearn import preprocessing 
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
X, y=make_regression(n_samples=100,n_features=1,noise=4,bias=100, random_state= 8888)
```

```python
plt.ylabel('X')
plt.xlabel('y')
plt.scatter(X,y, marker='*',color='r')
plt.show()
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3, random_state = 101)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
prd = model.predict(X)
plt.scatter(X,y, marker='*',color='r')
plt.plot(X,prd ,color='green')
plt.show()
```

```python
df=df.dropna()
df
```

```python
X1=df.iloc[:,1:]
y1=df[5]
X1_train, X1_test, y1_train, y1_test= train_test_split(X1, y1, test_size= 0.3)
```

```python
model = LinearRegression()
model.fit(X1_train, y1_train)
```

```python
test_prediction1 = model.predict(X1_test)
```

# TASK4

```python
s_r_sq=model.score(X1_train, y1_train)
print('coefficient of determination:', s_r_sq ,'\n')
```

```python
MSE1 = mean_squared_error (y1_test, test_prediction1)
RMSE1 = np.sqrt(MSE1)
print(MSE1)
```
