# Task1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

from sklearn import preprocessing 
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('DS2_C5_S2_Mtcars_Data_Practice.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0',axis=1)
```

```python
fig,  axes = plt.subplots(nrows= 1, ncols=6, figsize=(20,7))
axes[0].plot(df['cyl'],df['mpg'],'o')
axes[1].plot(df['disp'],df['mpg'],'o')
axes[2].plot(df['hp'],df['mpg'],'o')
axes[3].plot(df['drat'],df['mpg'],'o')
axes[4].plot(df['wt'],df['mpg'],'o')
axes[5].plot(df['qsec'],df['mpg'],'o')
plt.show()
```

```python
sns.pairplot(df, diag_kind='kde')
```

# Task2

```python
X = df[['disp']]
y= df['mpg']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3, random_state = 101)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
test_prediction = model.predict(X_test)
```

```python
s_r_sq=model.score(X_train, y_train)
print('coefficient of determination:', s_r_sq ,'\n')
print('intercept:', model.intercept_,'\n')
print('slope:', model.coef_)
```

```python
MAE = mean_absolute_error (y_test, test_prediction)
MSE = mean_squared_error (y_test, test_prediction)
RMSE = np.sqrt(MSE)
```

```python
print(MAE,MSE,RMSE)
```

```python
prd = model.predict(X)
```

```python
plt.ylabel('MPG')
plt.xlabel('disp')
plt.scatter(df['disp'],df['mpg'], marker='*',color='r')
plt.plot(df['disp'],prd ,color='green')
plt.show()
```
