# TASK4

```python
lst=[]
for i in range(len(review)):
    if re.search('magnitude.*',review[i].lower()):
        lst.append(i)
t4 = df.iloc[lst]
t4.shape
```

```python
t4.iloc[:,[2,3]]
```

```python
l1 = []
for i in t4['text']:
    l1.append(re.findall('\d+\.\d+',i))
s = []#sum of inside list
for i in l1:
    k = 0
    for j in i:
        k+= float(j)
    s.append(k)
sum1 = 0
for i in s:
    sum1+=i
average = sum1/len(s) 
print('Average is',average)
```
