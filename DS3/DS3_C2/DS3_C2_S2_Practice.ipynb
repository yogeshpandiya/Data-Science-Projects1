```python
import spacy
```

```python
nlp = spacy.blank("en")
```

```python
import re   # removing wide spaces using regex
d1 = re.sub('\s+',' ',d1)
```

```python
doc = nlp(d1)
```

```python
for token in doc[:100]:
    print(token)
```

# TASK2

```python
from collections import Counter

count = Counter()

for token in doc:
    count[token.orth_]+=1
```

```python
count.most_common(10)
```

```python
count = Counter()

for i in doc[:1500]:
    if i.orth_ == 'prisoners':
        count[i.orth_]+=1
    elif i.orth_ == 'prisoner':
        count[i.orth_]+=1
print(count)
```

```python
count = Counter()
doc1 = nlp(d1)
doc2 = nlp(d2)
doc3 = nlp(d3)
for i in doc1:
    if i.orth_.lower() == 'the':
        count['First file 48hrs']+=1
for i in doc2:
    if i.orth_.lower() == 'the':
        count['Second file Scifiscripts']+=1
for i in doc3:
    if i.orth_.lower() == 'the':
        count['Third file TenThingsIHate']+=1
print('count of (the) in different files',count)
```
