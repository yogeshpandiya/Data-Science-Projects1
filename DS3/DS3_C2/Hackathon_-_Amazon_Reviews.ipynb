```python
STOP_WORDS.update(["ingredients", "live", "person", "went", "came", "gave", "seen", "movie", '__label__1', '__label__2','label__1','label__2'])
```

```python
STOP_WORDS
```

```python
cleaned_text = df2[0].apply(lambda x: ' '.join(x for x in x.split() if x not in STOP_WORDS))
```

```python
cleaned_text
```

```python
import re
cleaned_text = cleaned_text.apply(lambda x: re.sub(r'[^\w\s]', '', x))
```

```python
cleaned_text
```

```python
nlp = spacy.load("en_core_web_sm")
#tokens = nlp(first_review)
token_text = cleaned_text.apply(lambda x: nlp(x))
```

```python
#token_text.apply(lambda y: i.lemma_ for i in y )
```

```python
token_text 
```

```python
import numpy as np
df2['POS Tag'] = np.nan
import warnings 
warnings.filterwarnings('ignore')
lst1 = []
for i in token_text:
    lst = []
    for j in i:
        lst.append(j.pos_)
    lst1.append(lst)
for i ,j in zip(lst1,range(910)):
    df2['POS Tag'][j]=' '.join(i) 
df2
```

```python
import warnings 
warnings.filterwarnings('ignore')
lst1 = []
for i in token_text:
    lst = []
    for j in i:
        lst.append(j.lemma_)
    lst1.append(lst)
for i ,j in zip(lst1,range(910)):
    df2[0][j]=' '.join(i)    
```
