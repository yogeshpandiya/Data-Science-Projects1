```python
import pandas as pd
import nltk
import spacy
from spacy.lang.en import English
nlp = spacy.load('en_core_web_sm')
```

```python
df = pd.read_csv('DS3_C2_S2_TravelReview_Data_Challenge.csv')
```

```python
df.isna().sum()
```

```python
review = df.content
review.head()
```

# TASK2

```python
nltk.word_tokenize(review[0])
```

```python
nltk_token = []
for i in review:
    nltk_token.append(nltk.word_tokenize(i))
```

```python
nltk_token 
```

#### tokenize using spacy 

```python
doc = nlp(review[0])
for token in doc:
    print(token)
```

```python
tokens = ''
for i in review:
    tokens+=i+' '
```

```python
doc = nlp(tokens[:10000]) # used limited words
for token in doc:
    print(token)
```
