```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
import pandas as pd
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay, classification_report
import numpy as np
```

```python
task1 = fetch_20newsgroups(subset='train')
```

```python
task1.target_names
```

# TASK2

```python
task2 = fetch_20newsgroups(subset='all', categories = ['alt.atheism', 'talk.religion.misc','comp.graphics','sci.space' ])
```

```python
dictn = {'Text': task2.data ,'Target': task2.target}
```

```python
df = pd.DataFrame(dictn)
```

```python
df.head()
```

```python
df.Target.replace(to_replace = [0,1,2,3] , value = task2.target_names, inplace = True)
```

```python
df.head()
```

```python
y = pd.get_dummies(df.Target)
y.head()
```

```python
X_train, x_test, Y_train, y_test = train_test_split(df.Text, y , test_size=0.2, random_state=12 )
```

```python
TFIDF = TfidfVectorizer()
```

```python
X_train_v = TFIDF.fit_transform(X_train).toarray()
```

```python
x_test_v = TFIDF.transform(x_test).toarray()
```

```python
X_train_v.shape
```

```python
x_test_v.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
mdl = LogisticRegression()
```

```python
from sklearn.multiclass import OneVsRestClassifier
```

```python
ovr = OneVsRestClassifier(mdl)
```

```python
ovr.fit(X_train_v, Y_train)
```

# TASK3

```python
y_pred = ovr.predict(x_test_v)
y_pred = pd.DataFrame(y_pred, columns = task2.target_names)
```

```python
pd.concat([y_pred.head(30).reset_index().add_prefix('Predict_'), y_test.head(30).reset_index().add_prefix('Actual_')], axis=1).drop(['Predict_index','Actual_index'],axis=1)
```

```python
print(classification_report(y_test,y_pred ))
```
