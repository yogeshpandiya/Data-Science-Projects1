# TASK2

```python
nltk.download('stopwords')
Stopwords = stopwords.words('english')
```

```python
line= open('challenge_train.txt','r').read() 
line = re.sub(r'[^\w\s]', '', line) # removing punctuation 
words = line.split()
for r in words:
    if not r in Stopwords:
        appendFile = open('cl_challenge_train.txt','a')
        appendFile.write(" "+r)
        appendFile.close()
```

## TASK3

```python
line= open('Review3.txt','r').read() 
line = re.sub(r'[^\w\s]', '', line) # removing punctuation 
words = line.split()
for r in words:
    if not r in Stopwords:
        appendFile = open('cleaned_Review3.txt','a')
        appendFile.write(" "+r)
        appendFile.close()
```

## TASK4

```python
model_sup = fasttext.train_supervised(input='cl_challenge_train.txt', lr=0.7, epoch=50, wordNgrams=3, bucket=20000)
```

## TASK5

```python
test_num, precision, recall = model_sup.test('cleaned_Review3.txt')
f_score = (2*precision*recall)/(precision+recall)
f_score
```

## TASK6

```python
check = []
for k in np.arange(1,10,1):
  for t in np.arange(0.1,1,0.1):
    test_num, precision, recall = model_sup.test('cleaned_Review3.txt', k=k, threshold=t)
    f_score = (2*precision*recall)/(precision+recall)
    check.append(f_score)

list1 = []
list2 = []

for k in np.arange(1,10,1):
  for t in np.arange(0.1,1,0.1):
    list1.append(k) 
    list2.append(t)
    
mydf = pd.DataFrame([list1,list2,check]).transpose().sort_values(2,ascending = False)
mydf.columns = ['K Value','Threshold value','f_score']
display(mydf.dropna())
```

## As we can see highest f_score: 0.002073 we are getting with K-Value: 5	  and Threshold-value: 0.3
