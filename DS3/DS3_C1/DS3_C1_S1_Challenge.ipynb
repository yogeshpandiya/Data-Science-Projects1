# TASK1

We will use the chi-square test of homoginity because the dependent variable is categorical.

degree of freedom = (c - 1) 

## a)  Set Null and Alternate Hypothesis

    H0: The distribution of regular applications accepted is the same as early applications accepted.
    Ha: The distribution of regular applications accepted is not the same as early applications accepted.

## b) Calculate critical Values

#### Set the value of alpha (α)

```python
alpha = 0.05 #considering alpha 
```

#### Table

```python
table = [[2115,1792,5306,1734,2685,1245],[577,627,1228,444,1195,761]]
```

#### Degree of Freedom

```python
df = 6-1 # number of columns - 1
df
```

```python
from scipy.stats import chi2
critical = chi2.ppf(1-alpha, df)
critical
```

## C) Calculate Statistics

```python
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(table)
stat
```

## d) Conclusion

```python
p_value = p
p
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,40,0.05)
plt.axvline(x=11.07,ls= '--',c='r',ymax=0.3)
plt.plot(x, chi2.pdf(x,df=df),c='g');
```

## Statistical Conclusion

### As per, decision rule:
    If p-value < alpha: Rejection of Null Hypothesis(H0)

    If ꭕ2-statistic > ꭕ2-critical : Rejection of Null Hypothesis(H0)
### From our study,
    p-value = 9.796462672909725e-9 < alpha = 0.05, which leads to rejection of the Null Hypothesis.

    ꭕ2-statistic = 430 > ꭕ2-critical = 11.070, which leads to rejection of Null Hypothesis.

## Decision

###    The two types of applications are not from the same distribution
