# TASK1

We will use the chi-square goodness of fit test as the expected distribution is a uniform distribution,

df = k – 1

## a)  Set Null and Alternate Hypothesis

* H0 = Frequency of observed data in this survey matches with expected frequency.
* Ha = Frequency of observed data in this survey is different with expected frequency.

## b) Calculate critical Values

#### Set the value of alpha (α)

```python
alpha = 0.05 #given in the question
```

#### Collect sample data

```python
data=[24, 108, 61, 14]
```

#### Degree of Freedom

```python
df = len(data)-1
df
```

#### Chi2 critical

```python
from scipy.stats import chi2

chi_critical = chi2.ppf(1-alpha, df=df)
print("Chi2-critical =",chi_critical)
```

## C) Calculate Statistics

#### Expected Frequency

```python
ef = sum(data)/len(data)
ef
```

#### Chi2 statistic

```python
chi2_s=0
for i in data:
    chi2_s+=(i-ef)**2/ef

print('chi2_statistic =',chi2_s)
```

## d) Conclusion 

#### P-Value

```python
from scipy.stats import norm

p_value = norm.sf(abs(chi2_s),df) #One tailed test
print("The p_value is "+str(p_value))
```

#### Graph

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,40,0.05)
plt.axvline(x=7.81,ls= '--',c='r',ymax=0.3)
plt.plot(x, chi2.pdf(x,df=df),c='g');
```

#### Statistical Conclusion

**As per, decision rule:**
* If p-value < alpha: Rejection of Null Hypothesis(H0)
* If ꭕ2-statistic > ꭕ2-critical : Rejection of Null Hypothesis(H0)

**From our study**
    
    p-value = 0.000 < alpha = 0.05, which leads to rejection of the Null Hypothesis.
    
    ꭕ2-statistic = 105.213 > ꭕ2-critical = 7.815, which leads to rejection of Null Hypothesis.

## Descision:

###  Both the rules leads to reject the null hypothesis hence observed frequencies are different from expected
